So, for annova, score f_ration = Difference between Groups / Within Group Difference, so for that Distplot can be a very good plot to visualize all that difference then calculate the annova

```python
def showDistribution(df,target,group):
    f, ax = plt.subplots(figsize=(10,6))
    Groups = df[group].unique()
    plt.title('Weight Distribution for {} on each {}'.format(target,group))
    
    for each in Groups[0:4]:
        sns.distplot(df[target][df[group]==each],label='{}'.format(each))
    
    plt.legend()
    plt.show()
```

```python
# Function designed in a custom way to replace categories in the ascending order of mean of the target column
def custom_encoding(Col, target, df):
    temp_df = df.groupby(Col)[target].mean().sort_values(ascending=True)
    temp_df = pd.DataFrame(temp_df)
    for i in range(len(temp_df)):
        temp_df['encoding'] = i+1
    for i in range(len(temp_df)):
        temp_df['encoding'][i] = i+1
    
    for i in range(len(temp_df)):
        temp_df['name'] = temp_df[temp_df['encoding']==i+1].index[0]
    
    for i in range(len(temp_df)):
        temp_df['name'][i] = temp_df[temp_df['encoding']==i+1].index[0]
        
    dict_temp = temp_df.set_index('name')['encoding'].to_dict()
    df = df.replace({Col:dict_temp})
    
    return df
```
